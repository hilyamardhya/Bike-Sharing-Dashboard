## Menentukan Pertanyaan Bisnis

- Bagaimana dampak keadaan cuaca terhadap total penyewaan sepeda?
- Apakah ada variasi dalam jumlah pengajuan sepeda antara hari biasa dan hari libur?

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
df = pd.read_csv("day.csv")
df.head()
```

**Insight:**
Dataset berhasil dimuat dan berisi informasi terkait waktu, cuaca, serta jumlah penyewaan sepeda.

### Assessing Data

```python
df.info()
df.isnull().sum()
df.describe()
```

**Insight:**
Tidak ditemukan missing value sehingga data dapat langsung digunakan untuk analisis lebih lanjut.

### Cleaning Data

```python
# Mengubah tipe tanggal
df['dteday'] = pd.to_datetime(df['dteday'])
```

**Insight:**
Kolom tanggal telah dikonversi ke format datetime untuk mempermudah analisis berbasis waktu.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
df.groupby("weathersit")["cnt"].mean()
df.groupby("workingday")["cnt"].mean()
```

**Insight:**
- Cuaca memiliki pengaruh terhadap jumlah penyewaan
- Hari kerja menunjukkan pola penyewaan yang lebih stabil

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
sns.barplot(x="weathersit", y="cnt", data=df)
plt.title("Pengaruh Cuaca terhadap Penyewaan Sepeda")
plt.xlabel("Kondisi Cuaca")
plt.ylabel("Jumlah Penyewaan")
plt.show()
```

### Pertanyaan 2:

```python
sns.barplot(x="workingday", y="cnt", data=df)
plt.title("Perbandingan Hari Kerja vs Libur")
plt.xlabel("Working Day (1=Ya, 0=Tidak)")
plt.ylabel("Jumlah Penyewaan")
plt.show()
```

**Insight:**
- Cuaca cerah menghasilkan jumlah penyewaan tertinggi, sedangkan kondisi cuaca buruk menurunkan jumlah penyewaan secara signifikan.
- Hari kerja memiliki tingkat penyewaan yang lebih stabil dibandingkan hari libur.

## Analisis Lanjutan (Opsional)

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Korelasi Antar Variabel")
plt.show()
```
